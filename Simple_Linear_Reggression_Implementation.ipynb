```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load dataset
data = {
    'Size': [1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400],
    'Price': [300000, 320000, 340000, 360000, 380000, 400000, 420000, 440000, 460000, 480000]
}
df=pd.DataFrame(data)
df
```

```python
#visualize data
plt.scatter(df['Size'], df['Price'])
plt.title('House Size vs price')
plt.xlabel('Size(sq ft)')
plt.ylabel('Price($)')
plt.show()
```

```python
#split data into training and test data
x=df[['Size']]
y=df[['Price']]
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)

#create a linear regression model
model=LinearRegression()

#train the model
model.fit(x_train, y_train)

#make a prediction
y_pred=model.predict(x_test)

#evalute a model
mse=mean_squared_error(y_test, y_pred)
r_sqaured=r2_score(y_test, y_pred)
print(f"Mean Squared error:{mse}")
print(f"R-Sqaured:{r_sqaured}")
```

```python
#vizyalising data
plt.figure(figsize=(8,6))

# Actual test data
plt.scatter(x_test, y_test, label='Actual Prices')

# Predicted test data
plt.scatter(x_test, y_pred, label='Predicted Prices')

# Regression line (training data)
plt.plot(x_train, model.predict(x_train), label='Regression Line')

plt.title("Linear Regression Model")
plt.xlabel("Size (sq ft)")
plt.ylabel("Price ($)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([1000,1500,2000,2500,3000]).reshape(-1,1)
y=np.array([300,400,500,600,700])

#create and train the model
model=LinearRegression()
model.fit(x,y)

# Get slope (m) and intercept (b)
m = model.coef_[0]
b = model.intercept_

print(f"Slope (m): {m:.2f}")
print(f"Intercept (b): {b:.2f}")
print(f"Equation: y = {m:.2f}x + {b:.2f}")

# Predictions
y_pred = model.predict(x)

# Plot
plt.figure(figsize=(8,6))

plt.scatter(x, y, label='Data points')
plt.plot(x, y_pred, label='Linear regression')

plt.xlabel('Size (sq ft)')
plt.ylabel('Price ($1000s)')
plt.title('Univariate Linear Regression: House Price vs Size')
plt.legend()
plt.show()

# Predict price for new house size (e.g., 2200 sq ft)
new_size = np.array([[2200]])
predicted_price = model.predict(new_size)

print(f"Predicted price for 2200 sq ft: ${predicted_price[0]:.2f}K")
```
